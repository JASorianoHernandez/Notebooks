### 2.4 Big O of k-nearest neighbors
Dentro de los parámetros que representan la eficiencia de un programa computacional, podemos definirla de acuerdo al tiempo de ejecución, en otras palabras, podemos observar lo siguiente:



Cabe destacar que depeniendo de la complejidad del modelo, al aplicarlo al data set, este puede verse limitado por el tiempo de ejecución, por lo cual mientras más bajo sea el tiempo de ejecución será mejor para el desempeño del programa.



```python
### ejemplo de big O
A = [[1, 2, 3],
     [2, 3, 4]]

B= [[5,8],
    [6,9],
    [7,10]]

def matrixmultiply(A, B):
    C= [[0 for row in range(len(A))]
       for col in range(len(B[0]))]
    
    for row_a in range (len(A)):
        for col_b in range (len(B[0])):
            for col_a in range(len(A[0])):
                C[row_a][col_b] += \
                    A[row_a][col_a]*B[col_a][col_b]
    return C

matrixmultiply(A,B)
```

En este caso podemos ver dos for anidados, una con dos ciclos for y otro con 3 ciclos for por lo tanto su complejidad computacional es de n al cubo

Big O of kNN
- Considerando K vecinos
- Se tiene que considerar el hecho de que se toma la distnacia de cada una de las muestras n del data set
- Se tiene que consdierar la distancia de acuerdo a la dimensión m del arreglo, esto afecta la complejidad del algoritmo

En este sentido la complejidad O sería  O(K* n * m) sin embargo los valores de k y m son mucho menores que n por lo tanto la complejidad quedaería solo como o ( n )
